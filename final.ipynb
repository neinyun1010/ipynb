```python
#若我們今天想衡量於某一特定區域或時間內剛好發生 x 次之機率: λ為給定時間或空間內發生次數

import numpy as np
from math import exp, factorial

class PoissonDistributionSimulation:
    def __init__(self, lambda_parameter, num_simulations):
        self.lambda_parameter = lambda_parameter 
        self.num_simulations = num_simulations
        self.simulated_data = []

    def simulate(self):
        self.simulated_data = np.random.poisson(self.lambda_parameter, self.num_simulations)
    def print_results(self):
        print("Simulated results:", self.simulated_data[:10])
        print("mean:", np.mean(self.simulated_data).round(2))
        print("variance:" , np.var(self.simulated_data).round(2))
    def calculate_poisson_pmf(self, x):
        pmf_value = (exp(-self.lambda_parameter) * self.lambda_parameter**x) / factorial(x)   
        return pmf_value

    def print_individual_poisson_pmf(self):
        print(f'Poisson Distribution PMF with λ = {self.lambda_parameter}') 
        print("Poisson PMF for the first 10 simulated samples:")
        for x in self.simulated_data[:10]:
            pmf_value = self.calculate_poisson_pmf(x)
            print(f'Simulated results {x:<10} Probability {pmf_value:.5f}')
```

```python
poisson_simulation = PoissonDistributionSimulation(lambda_parameter=5, num_simulations=100)

poisson_simulation.simulate()

poisson_simulation.print_results()

poisson_simulation.print_individual_poisson_pmf()
```

```python
import matplotlib.pyplot as plt

def plot_poisson_distribution(self):
        plt.hist(self.simulated_data, bins=np.arange(min(self.simulated_data), max(self.simulated_data) + 1)- 0.5 , density=True, label='Simulated Data')
        x_values = np.arange(0, max(self.simulated_data) + 1) 
        pmf_values = [self.calculate_poisson_pmf(x) for x in x_values] 
        
        plt.plot(x_values, pmf_values, 'ro-', label='Poisson PMF')
        plt.title(f'Poisson Distribution Simulation (λ = {self.lambda_parameter})')
        plt.xlabel('Number of Events')
        plt.ylabel('Probability')
        plt.legend()
        plt.show()

plot_poisson_distribution(poisson_simulation)
```
